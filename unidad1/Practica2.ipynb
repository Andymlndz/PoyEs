```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
    numero_aleatorio = round(random.uniform(50, 100), 2)
    fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []

for i in range(100):
    numero_aleatorio2 = round(random.uniform(40, 93), 2)
    fabrica2.append(numero_aleatorio2)

print(fabrica2)
```

```python
maximo1 = max(fabrica1)         # Valor máximo
minimo1 = min(fabrica1)         # Valor mínimo
rango1 = maximo1 - minimo1      # Rango (diferencia entre máximo y mínimo)
media1 = mean(fabrica1)         # Media aritmética
mediana1 = median(fabrica1)     # Mediana
moda1 = mode(fabrica1)          # Moda

# Impresión de los resultados
print("Estadística descriptiva de la Fábrica 1\n",
      "Media =", media1,
      "\nMediana =", mediana1,
      "\nModa =", moda1,
      "\nMáximo =", maximo1,
      "\nMínimo =", minimo1,
      "\nRango =", rango1)
```

```python
maximo2 = max(fabrica2)         # Valor máximo
minimo2 = min(fabrica2)         # Valor mínimo
rango2 = maximo2 - minimo2      # Rango corregido (antes usaba maximo1 por error)
media2 = mean(fabrica2)         # Media aritmética
mediana2 = median(fabrica2)     # Mediana
moda2 = mode(fabrica2)          # Moda

# Impresión de los resultados
print("Estadística descriptiva de la Fábrica 2\n",
      "Media =", media2,
      "\nMediana =", mediana2,
      "\nModa =", moda2,
      "\nMáximo =", maximo2,
      "\nMínimo =", minimo2,
      "\nRango =", rango2)
```

```python
desviacion_estandar_fabrica1 = stdev(fabrica1)
print(desviacion_estandar_fabrica1)
```

```python
desviacion_estandar_fabrica2 = stdev(fabrica2)
print(desviacion_estandar_fabrica2)
```

```python
plt.figure(figsize=(8,6))
plt.hist(fabrica2, bins=10, alpha=0.2, color='b', edgecolor='black')
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label=f'Media={media2:.2f}')
plt.axvline(mediana2, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana2:.2f}')
plt.axvline(media2 + desviacion_estandar_fabrica2, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estándar = {desviacion_estandar_fabrica2:.2f}')
plt.axvline(media2 - desviacion_estandar_fabrica2, color='black', linestyle='dashed', linewidth=2)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Producción diaria de la fabrica 2')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.hist(fabrica1, bins=10, alpha=0.2, color="b", edgecolor="black")
plt.axvline(mediana1, color="r", linestyle="dashed", linewidth=2, label=f"Mediana={mediana1:.2f}")
plt.axvline(moda1, color="y", linestyle="dashed", linewidth=2, label=f"Moda={moda1:.2f}")
plt.axvline(mediana1 - desviacion_estandar_fabrica1, color="black", linestyle="dashed", linewidth=2, label=f"Desv. Estándar = {desviacion_estandar_fabrica1:.2f}")
plt.axvline(mediana1 + desviacion_estandar_fabrica1, color="black", linestyle="dashed", linewidth=2)
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.title("Producción diaria de la fabrica 1")
plt.legend()
plt.show()
```
